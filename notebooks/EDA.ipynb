Graficos

```python
avg_price_df = df.groupby('model', as_index=False)['price'].mean()
```

```python
top_20_costosos = avg_price_df.sort_values(by='price', ascending=False).head(20)
```

```python
fig = px.bar(top_20_costosos, x='model', y='price',
              title='Los 20 Modelos Más Costosos',
              color='price',  
              color_continuous_scale='Viridis')


fig.update_layout(xaxis_title='Modelo', yaxis_title='Precio Promedio',
                  xaxis_tickangle=-45, 
                  plot_bgcolor='rgba(0, 0, 0, 0)',  
                  title_x=0.5) 


fig.show()
```

```python
avg_day_df = df.groupby('model', as_index=False)['days_listed'].mean()
top_20_tiempo_ventas = avg_day_df.sort_values(by='days_listed', ascending=False).head(20)
```

```python
fig = px.bar(avg_day_df.sort_values(by='days_listed', ascending=False).head(20),
              x='model', y='days_listed',
              title='Los 20 Modelos Más Difíciles de Vender',
              color='days_listed',
              color_continuous_scale='Inferno',  # Nueva escala de colores
              text='days_listed',  # Muestra los valores en las barras
              labels={'days_listed': 'Días Listados', 'model': 'Modelo'})
fig.update_layout(xaxis_title='Modelo', yaxis_title='Días Listados',
                  xaxis_tickangle=-45,  # Rotar etiquetas para mejor visibilidad
                  plot_bgcolor='rgba(255, 255, 255, 0)',  # Fondo blanco
                  title_x=0.5)

fig.show()
```
